## 전기차 현황

```python
file = pd.read_excel('../data/자동차+등록현황(연료별).xlsx')
```

```python
file
file.set_index('차종별(2)', inplace=True)
year = ['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']
```

```python
file_international.info()
file_international = file_international.astype({'시점' : str})
```

```python
pattern = "\d+"
file_international['시점(연도)'] = file_international['시점'].apply(lambda x:re.findall(pattern, x))
file_international['시점(연도)'] = file_international['시점(연도)'].apply(lambda x:x[0])
```

```python
file_international.columns
```

```python
sum_international = file_international.loc[:, ['운항 (편), 계', '운항 (편), 도착', '운항 (편), 출발', '여객 (명), 계',
       '여객 (명), 도착', '여객 (명), 출발', '화물 (톤), 계', '화물 (톤), 도착', '화물 (톤), 출발',
       '시점(연도)']].groupby('시점(연도)').sum()
```

```python
color = ['#459395', '#eb7c69', '#fda638']
sum_domestic
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

year = list(sum_domestic.index)
cols = list(sum_domestic.columns)
# year
ax.set_facecolor('white')

for i, c in enumerate(cols) :
    ax = plt.plot(year, sum_domestic.loc[:, c], color = color[i], label = c, linewidth = 3)
    if '여객' in c:
        for idx, v in enumerate(sum_domestic.loc[:, c]):
            plt.text(idx, v, format(int(v), ','), ha = 'center')

current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])

plt.legend()
plt.title('국내선 통계', fontsize = 14)
plt.show()
```
